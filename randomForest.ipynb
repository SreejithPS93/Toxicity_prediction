```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/data2.csv',)
df.head()
```

```python
df.columns
```

```python
mol_data = df.drop(['Unnamed: 0','TAID', 'Pubchem CID', 'IUPAC Name', 'SMILES',
       'Canonical SMILES', 'InChIKey', 'mouse_intraperitoneal_LD50'],axis = 1)
```

```python
mol_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df['mouse_intraperitoneal_LD50'] = scaler.fit_transform(df[['mouse_intraperitoneal_LD50']])
```

```python
df.head()
```

```python
from sklearn.model_selection import tra
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(mol_data,df['mouse_intraperitoneal_LD50'],test_size = 0.2,random_state = 42)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf_model.fit(X_train,Y_train)
```

```python
rf_pred = rf_model.predict(X_test)
```

```python
mse_forest = mean_squared_error(Y_test,rf_pred)
r2_forest = r2_score(Y_test,rf_pred)
print(f'Mean Squared Error : {mse_forest}, R-squared: {r2_forest}')
```
